```python
import numpy as np

import matplotlib.pyplot as plt
import time


# Função que executa O(n^2)
def quadratic(n):
    result = 0
    for i in range(n):
        for j in range(n):
            result += i * j
    return result


# Lista com os valores de n
n_values = list(range(1, 600))
# print(n_values)
# Lista com os tempos de execução correspondentes
times = []

# Loop para medir o tempo de execução para cada v
# alor de n
for n in n_values:
    start_time = time.time()
    quadratic(n)
    end_time = time.time()
    times.append(end_time - start_time)

# Plotagem do gráfico
plt.plot(n_values, times)
plt.xlabel('n')
plt.ylabel('Tempo de execução (s)')
plt.show()
```
