```python
import time
import math
import random
import numpy as np
```

```python
def get_first_element(lst):
    return lst[0]

tempos_get_first_element = []
tamanhos = range(10,10000,20)
for tam in tamanhos:
    lista = random.sample(range(1, 100000), tam)
    tamanho = len(lista)
    start = time.time()
    get_first_element(lista)
    tempos_get_first_element.append(time.time() - start)


np.savetxt('tempos_get_first_element.txt', tempos_get_first_element)
```

```python

def merge_sort(lst):
    if len(lst) <= 1:
        return lst
    mid = len(lst) // 2
    left = lst[:mid]
    right = lst[mid:]
    left = merge_sort(left)
    right = merge_sort(right)
    return merge(left, right)

def merge(left, right):
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result += left[i:]
    result += right[j:]
    return result

tempos_merge_sort = []
tamanhos = range(10,1000000,10)
for tam in tamanhos:
    lista = random.sample(range(1, 1000000), tam)
    tamanho = len(lista)
    start = time.time()
    merge_sort(lista)
    tempos_merge_sort.append(time.time() - start)

np.savetxt('tempos_merge_sort.txt', tempos_merge_sort)
```

```python
def bubble_sort(lst):
    for i in range(len(lst)):
        for j in range(len(lst) - 1):
            if lst[j] > lst[j + 1]:
                lst[j], lst[j+1] = lst[j+1], lst[j]
    return lst


tempos_bubble_sort = []
tamanhos = range(10,10000,20)
for tam in tamanhos:
    lista = random.sample(range(1, 100000), tam)
    tamanho = len(lista)
    start = time.time()
    get_first_element(lista)
    tempos_get_first_element.append(time.time() - start)
    start = time.time()
    merge_sort(lista)
    tempos_merge_sort.append(time.time() - start)
    start = time.time()
    bubble_sort(lista)
    tempos_bubble_sort.append(time.time() - start)

np.savetxt('tempos_bubble_sort.txt', tempos_bubble_sort)
```

```python
import matplotlib.pyplot as plt

tempos_get_first_element = np.loadtxt('get_first_element.txt')
tempos_merge_sort = np.loadtxt('merge_sort.txt')
tempos_bubble_sort = np.loadtxt('bubble_sort.txt')
tamanhos = np.loadtxt('tamanhos.txt')

plt.plot(tamanhos, tempos_get_first_element, label='O(1)')
plt.plot(tamanhos, tempos_merge_sort, label='O(n log n)')
# plt.plot(tamanhos, tempos_bubble_sort, label='O(n^2)')
# plt.plot(tamanhos, tempos_fib, label='O(n^3)')
plt.xlabel('Tamanho da entrada (n)')
plt.ylabel('Tempo de execuo (s)')
plt.title('Comparao de complexidade de algoritmos')
plt.legend()
plt.show()
```
