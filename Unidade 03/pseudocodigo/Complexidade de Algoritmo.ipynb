```python
# O(1)
def get_first_element(lst):
    return lst[0]
```

```python
# O(log n)
def binary_search(lst, item):
    low = 0
    high = len(lst) - 1
    while low <= high:
        mid = (low + high) // 2
        guess = lst[mid]
        if guess == item:
            return mid
        if guess > item:
            high = mid - 1
        else:
            low = mid + 1
    return -1


```

```python
# O(n)
def search_list(lst, item):
    for i in range(len(lst)):
        if lst[i] == item:
            return i
    return -1
```

```python
# O(n log n)
# def quick_sort(lst):
#     if len(lst) < 2:
#         return lst
#     else:
#         pivot = lst[0]
#         less = [i for i in lst[1:] if i <= pivot]
#         greater = [i for i in lst[1:] if i > pivot]
#         return quick_sort(less) + [pivot] + quick_sort(greater)

def merge_sort(lst):
    if len(lst) <= 1:
        return lst
    mid = len(lst) // 2
    left = lst[:mid]
    right = lst[mid:]
    left = merge_sort(left)
    right = merge_sort(right)
    return merge(left, right)

def merge(left, right):
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result += left[i:]
    result += right[j:]
    return result
```

```python
# O(n^2)
def bubble_sort(lst):
    for i in range(len(lst)):
        for j in range(len(lst) - 1):
            if lst[j] > lst[j + 1]:
                lst[j], lst[j+1] = lst[j+1], lst[j]
    return lst
```

```python
# O(2^n) recebendo uma lista
def fibonacci_list(lst):
    result = []
    for num in lst:
        result.append(fibonacci(num))
    return result

def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

```
